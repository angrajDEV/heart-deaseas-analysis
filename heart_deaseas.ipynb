```python
import os
import pandas as pd 
import seaborn as sns

data = pd.read_csv('heart.csv')
data.head()
```

-----------------------------------------------------------   age , distplot  ---------------------------------------------------------------

```python
sns.displot(x=data['Age'] , kde='True' ,color='blue' )
```

```python
sns.displot(x=data['RestingBP'],kde='True',color='blue')
```

```python
sns.displot(data['Cholesterol'] , kde='True' , color='blue')
```

```python
data.groupby('RestingECG').size().plot(kind='pie' , autopct='%.1f')
```
